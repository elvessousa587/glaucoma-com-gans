## Importando as bibiliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from glob import glob
from skimage.io import imread, imsave,imread_collection
from sklearn.cluster import KMeans
from skimage.color import rgb2gray
from scipy import ndimage as ndi
# filters de bordas
from skimage.filters import sobel, roberts, prewitt, scharr, threshold_otsu
from skimage.filters import try_all_threshold
from skimage.color import rgb2gray
```

## Lendo as imagens originais e fazendo o Plote

```python
imagens = imread_collection('testando/teste/RIMONE-db-r2/Normal/*.jpg')
```

```python

#Imagens Originais
fig, ax = plt.subplots(1, 6, figsize=(10,5)) 
for i, img in enumerate(imagens):
    ax[i].imshow(img, cmap='gray')
    ax[i].set_title('imagem ' + str(i + 1))
```

```python
#Converto as imagens para tons de cinza, mais precisamente o canal green e salvo e leio as imagens salvas

for i, img in enumerate(imagens):
    green = img[:,:,1]
    cinza=rgb2gray(green)
    imsave('testando/imagensCinza/img'+(str(i))+'.jpg', cinza)
    
    
#leio as imgs cinzas
imagensC = imread_collection('testando/imagensCinza/*.jpg')
```

```python
#Imagens Cinzas
fig, ax = plt.subplots(1, 6, figsize=(10,5)) 
for i, img in enumerate(imagensC):
    ax[i].imshow(img, cmap='gray')
    ax[i].set_title('imagem ' + str(i + 1))
```

## Segmentação das imagens com (K-Means)

```python
# faço um for percorrendo todas imagens e após segmentar todas elas salvo as imagens segmentadas
# na pasta imagens_seg
for i, img in enumerate(imagensC):
    model_kmeans = KMeans(n_clusters=2, random_state=0).fit(np.reshape(img,(-1, 1)))
    segmented1 = model_kmeans.predict(np.reshape(img,(-1, 1)))
    segmented2 = np.reshape(segmented1,(img.shape[0],img.shape[1]))
    imsave('testando/segmentadas/img_seg'+(str(i))+'.jpg', segmented2)
                                                        
```

## Função para verificar as quantidade dos pixel preto e branco

```python
def verifica_quantidade_de_pixel_preto_e_branco(image):
    qtde_pixel_preto = 0
    qtde_pixel_branco = 0
    for linha in range(image.shape[0]):
        for coluna in range(image.shape[1]):
            #for c in range(image.shape[2]):
                if image[linha][coluna] == 0:
                    qtde_pixel_preto += 1
                else:
                    qtde_pixel_branco += 1
    return qtde_pixel_preto, qtde_pixel_branco
```

## Função que verifica se a segmentação ficou com o núcle da imagem toda preta, caso contrario muda os pixel preto para branco.

```python
# verificando se a imagem ficou com segmentação preta
# funcao para mudar a cora da imagem
def altera_pixel_preto_e_branco(imagens):
    for i, image in enumerate(imagens):
        
        # chamaa a função verifica_quantidade_de_pixel_preto_e_branco
        # passando como parametro a imagem
        # atribuindo o retorno da qtde de pixel preto e branco daquela imagem
        qtde_pixel_preto, qtde_pixel_branco = verifica_quantidade_de_pixel_preto_e_branco(image)
        
        # verifico se a qtde de pixel preto e maior que a qtde de pixel branco e faco a substituição
        if qtde_pixel_preto > qtde_pixel_branco:
            # Altera os elementos de uma matriz com base em valores condicionais os de entrada.
            np.place(image, image == 0, [255])
            imsave('testando/segmentadas2/imagem_seg'+(str(i))+'.jpg', image)
        else:
            imsave('testando/segmentadas2/imagem_seg'+(str(i))+'.jpg', image)
```

## Chamo a função passando a lista das imagens segmentadas para corrigir as imagens que ficaram o núcle preto e não branco


```python
# passando a lista de imagem a ser verificada para se o centro da imagem for preto
# fazer a alteração pelo pixel branco [255]
# passando a lista de imagens como parametro para a funcao (folhas_seg)

altera_pixel_preto_e_branco(imagens_seg)
```
